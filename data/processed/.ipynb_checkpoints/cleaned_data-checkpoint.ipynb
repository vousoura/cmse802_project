```python
import pandas as pd
```

```python
raw_data = pd.read_excel("raw data.xlsx") # Load the raw dataset from an Excel file
raw_data.head() ## Show the first few rows to check the structure of the dataset
```

```python
cleaned_data =raw_data.dropna() # Drop rows with missing values to ensure my data is clean
cleaned_data.head(15) # Show the first 15 rows of the cleaned dataset
```

```python
slip_planes_of_interest = [110, 112, 123] #I filtered the data to have only 3 kinds of planes; (110),(112),(123)
filtered_data = cleaned_data[cleaned_data.iloc[:, 1].isin(slip_planes_of_interest)] # Filter the dataset to include only the rows with the planes I want
filtered_data = filtered_data.rename(columns={"W": "Solute_Concentration"}) # Rename the column "W" to "Solute_Concentration" for better understanding
filtered_data.tail(15) # Show the last 15 rows of the final filtered dataset
```
